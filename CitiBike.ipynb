```python
# Add dependencies
import glob
import pandas as pd
```

```python
# glob.glob('data*.csv') - returns List[str]
# pd.read_csv(f) - returns pd.DataFrame()
# for f in glob.glob() - returns a List[DataFrames]
# pd.concat() - returns one pd.DataFrame()
CitiBike_data_ridership = pd.concat([pd.read_csv(f) for f in glob.glob('Resources/CitiBikeDailyRidershipandMembershipData/*.csv')], ignore_index = True)
CitiBike_data_trip = pd.concat([pd.read_csv(f) for f in glob.glob('Resources/CitiBikeTripHistories/*.csv')], ignore_index = True)
```
